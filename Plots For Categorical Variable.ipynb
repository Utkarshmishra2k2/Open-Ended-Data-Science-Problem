Multiple Sets Of Plots

```python

sns.kdeplot(Penguin.body_mass_g)   
sns.boxplot(Penguin.body_mass_g)  
sns.swarmplot(Penguin.body_mass_g)  
sns.violinplot(Penguin.body_mass_g) 
```

```python
fig,axs=plt.subplots(nrows=4) 
sns.kdeplot(Penguin.body_mass_g,ax=axs[0]);   
sns.boxplot(Penguin.body_mass_g,ax=axs[1]);  
sns.swarmplot(Penguin.body_mass_g,ax=axs[2]); 
sns.violinplot(Penguin.body_mass_g,ax=axs[3]); 
```

```python
#Here I Ploted The All Graph In Different Graph With Size
fig,axs=plt.subplots(nrows=4)       
fig.set_size_inches(10,15);        
sns.kdeplot(Penguin.body_mass_g,ax=axs[0]);  
sns.boxplot(Penguin.body_mass_g,ax=axs[1]);    
sns.swarmplot(Penguin.body_mass_g,ax=axs[2]); 
sns.violinplot(Penguin.body_mass_g,ax=axs[3]); 
```

```python
#Here I Ploted The All Graph In Different Graph With With Graph Axis Limts
fig,axs=plt.subplots(nrows=4)        
fig.set_size_inches(10,15);          
a=sns.kdeplot(Penguin.body_mass_g,ax=axs[0]);   
a.set(xlim=(1500,7500))                          
b=sns.boxplot(Penguin.body_mass_g,ax=axs[1]);    
b.set(xlim=(1500,7500))                          
c=sns.swarmplot(Penguin.body_mass_g,ax=axs[2]);  
c.set(xlim=(1500,7500))                         
d=sns.violinplot(Penguin.body_mass_g,ax=axs[3]);
d.set(xlim=(1500,7500))                          
```

```python
#Here I Ploted The All Graph In Different Graph With Size
fig,axs=plt.subplots(nrows=4)       
fig.set_size_inches(10,15);          
a=sns.kdeplot(Diamonds.price,ax=axs[0]);   
a.set(xlim=(1500,7500))                    
b=sns.boxplot(Diamonds.price,ax=axs[1]);    
b.set(xlim=(1500,7500))                    
#c=sns.swarmplot(Diamonds.price,ax=axs[2]); 
#c.set(xlim=(1500,7500))                     
d=sns.violinplot(Diamonds.price,ax=axs[3]);
d.set(xlim=(1500,7500))                     
```
